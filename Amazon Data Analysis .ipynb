# Aggregating Data

## Within this part I am going to find answers of these questions:-

#### The Total Sales by Category.

#### Average Amount by Category and Status.

#### Total Sales by Fulfilment and Shipment Type.

```python
# Total Sales by Category
category_totals = sales_data.groupby('Category') # in parenthesis here i am grouping data by category

category_totals = sales_data.groupby('Category')['Amount'] # After grouping the data by category,
                                                           # in squared brackets i have to take a column
                                                           # onto i need to perform sum/summing Amount

# After this i need to tell python that i needed to group by data by category and need to perform some sort of
# aggregation on 'Amount'.
# What method of aggregation ?
# I want to perform a sum.

category_totals = sales_data.groupby('Category')['Amount'].sum()
category_totals
# In sales data we are going to do groupby(category) and sum the 'Amount'
```

```python
# Now i want to add a index to the data
category_totals = sales_data.groupby('Category',as_index = False)['Amount'].sum()
category_totals
```

```python
# Now i want to sort the category_totals data into descending order
category_totals = category_totals.sort_values( 'Amount' , ascending = False )
category_totals
```

# Calculating the average amount by category and fulfillment

```python


fulfilment_average = sales_data.groupby(['Category','Fulfilment'])['Amount'].mean()
fulfilment_average
```

```python
# Adding Index to the fulfilment average
fulfilment_average = sales_data.groupby(['Category','Fulfilment'], as_index = False)['Amount'].mean()
fulfilment_average
```

```python
# Sorting fulfilment_average
fulfilment_average = fulfilment_average.sort_values('Amount',ascending = False)
fulfilment_average
```

# Average Amount by Category and Status.

```python
status_average = sales_data.groupby(['Category','Status'])['Amount'].mean()
status_average
```

```python
# Adding index to status_average
status_average = sales_data.groupby(['Category','Status'],as_index=False)['Amount'].mean()
status_average
```

```python
# Sorting the status_average
status_average = status_average.sort_values('Amount',ascending = False)
status_average
```

# Total Sales by Fulfilment and shipment type

```python
total_sales_ship = sales_data.groupby(['Courier Status','Fulfilment'])['Amount'].sum()
total_sales_ship
```

```python
# Adding Index
total_sales_ship = sales_data.groupby(['Courier Status','Fulfilment'],as_index = False)['Amount'].sum()
total_sales_ship
```

```python
# Sorting according to the sales
total_sales_ship = total_sales_ship.sort_values('Amount',ascending=False)
total_sales_ship
```
